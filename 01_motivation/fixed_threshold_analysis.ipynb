## 1. Clinical Context

Peptide Receptor Radionuclide Therapy (PRRT) with [¹⁷⁷Lu]-DOTATATE is an established treatment for metastatic neuroendocrine tumors. Personalized dosimetry — estimating the absorbed dose delivered to each tumor — is essential for treatment adaptation and outcome prediction.

Dosimetry accuracy depends entirely on the accuracy of **activity quantification**, which in turn depends on correct **tumor segmentation**. An incorrect segmentation boundary changes the volume over which mean activity concentration is computed, directly propagating into the dose estimate.

Camera calibration in quantitative SPECT relies on **Recovery Coefficient (RC) curves** derived from phantom acquisitions with known sphere sizes and activities. These RC curves correct for partial volume effects and are the backbone of absolute activity quantification.

## 2. CT Segmentation Limitations

In phantom calibration, CT-based delineation serves as the gold standard: sphere boundaries are sharp, geometry is known exactly, and CT contrast is unambiguous.

In clinical practice after PRRT, this is not always the case. Tumor visibility on CT depends on lesion type, location, and treatment response. Many lesions — particularly liver metastases after therapy — become isodense with surrounding tissue and **cannot be reliably delineated on CT**. In these cases, segmentation must be performed directly on the SPECT image.

SPECT-based segmentation introduces a fundamental question: **which threshold to apply?** Unlike CT, SPECT images are blurred by the scanner point spread function (PSF), and lesion boundaries depend on both lesion size and the local background activity level.

## 3. The 41% Threshold and Its Limitations

The most widely adopted solution in clinical SPECT dosimetry is the **fixed 41% threshold**: segment the lesion as all voxels above 41% of the maximum voxel intensity within the region of interest. This value was derived empirically from early phantom studies and has become a de facto standard.

The physical justification breaks down under two conditions that are common in clinical PRRT:

**Lesion size effects.** Partial volume effects scale with the ratio of lesion diameter to scanner resolution (FWHM ≈ 9–12 mm for modern SPECT/CT). Small lesions lose a larger fraction of their counts to the surrounding voxels. The threshold that correctly recovers mean activity is therefore higher for small lesions than for large ones.

**Background activity effects.** Background activity spills into the lesion boundary region, elevating apparent counts near the edge. Higher background-to-tumor ratios push the optimal threshold upward. A fixed value cannot account for this variation across patients and lesions.

The consequence is systematic quantification error whose direction and magnitude depend on lesion size and contrast — making it unpredictable without phantom characterization.

## 4. Toward a Better Approach

The data above establishes a clear need: the optimal segmentation threshold is not a constant. It depends on measurable physical properties of the lesion and its environment — specifically **lesion size** and **background-to-tumor activity ratio**.

Both quantities can be estimated directly from the SPECT image prior to final segmentation. This opens the possibility of a standardized, physics-informed rule that replaces the fixed 41% with a **variable threshold adapted to each lesion**.
