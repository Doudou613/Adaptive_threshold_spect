# Fixed Threshold Analysis


## 1. Clinical Context

Peptide Receptor Radionuclide Therapy (PRRT) with [¹⁷⁷Lu]-DOTATATE is an established treatment for metastatic neuroendocrine tumors. 

Estimating the absorbed dose delivered to each tumor (dosimetry) is essential for treatment adaptation and outcome prediction. Dosimetry accuracy depends entirely on the accuracy of **activity quantification**, which in turn depends on correct **tumor segmentation**. An incorrect segmentation boundary changes the volume over which mean activity concentration is computed, directly propagating into the dose estimate.

Camera calibration in quantitative SPECT relies on **Recovery Coefficient (RC) curves** which provide the correction factor needed to convert SPECT counts to true activity concentration, derived from phantom acquisitions where each sphere is delineated on CT.

## 2. CT Segmentation Limitations

In phantom calibration, CT-based delineation serves as the gold standard: sphere boundaries are sharp, geometry is known exactly, and CT contrast is unambiguous.

In clinical practice after PRRT, this is not always the case: tumor visibility on CT depends on lesion type, location, and treatment response. Many lesions become isodense with surrounding tissue and **cannot be reliably delineated on CT**.

In these cases, segmentation must be performed directly on the SPECT image by applying an intensity threshold, the choice of which depends on both lesion size and the local background activity level. Which leads to the question: **which threshold to apply?**

## 3. The 41% Threshold and Its Limitations

The most widely adopted solution in clinical SPECT dosimetry is the **fixed 41% threshold**: segment the lesion as all voxels above 41% of the maximum voxel intensity within the region of interest. This value was derived empirically from early phantom studies and has become a de facto standard.

Two physical effects govern the optimal threshold: **partial volume effects**, which cause small lesions to lose counts to surrounding voxels due to the finite scanner resolution, and **background spillover**, where activity from surrounding tissue elevates apparent counts near the lesion boundary.

Both effects vary across patients, lesions, and imaging conditions. A fixed threshold cannot account for this variability, leading to systematic quantification errors.

![Activity errors 41% vs CT segmentation](images/image-02.png)

The plot shows activity concentration errors when using the 41% threshold compared to CT-based segmentation across four sphere sizes. Errors reach up to ~47% for the 17 mm sphere and ~20% for the 28 mm sphere.

## 4. Toward a Better Approach

The data above establishes a clear need: the optimal segmentation threshold is not a constant. It depends on measurable physical properties of the lesion and its environment, specifically **lesion size** and **background-to-tumor activity ratio**.

Both quantities can be estimated directly from the SPECT image prior to final segmentation. This opens the possibility of a standardized, physics-informed rule that replaces the fixed 41% with a variable threshold adapted to each lesion.
